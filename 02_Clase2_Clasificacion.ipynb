# **¿Qué es una señal?**

Una señal es una función de una o más variables independientes que contiene información acerca del comportamiento o naturaleza de algún fenómeno físico.

# **¿Qué es un sistema?** 

Un sistema es un conjunto de elementos que responden a señales particulares generando otras señales o produciendo algún comportamiento en el mismo. 



**A continuación se realizará la clasificación de las señales dependiendo de su variable independiente, de su variable dependiente, de su máximo y por último de su duración.**

# **1. En función de su variable independiente / según su dominio**


**1.1 Señales continuas**

Las señales continuas se encuentran definidas para todos los instantes de tiempo y pueden modelarse como funciones que tienen como dominio los reales.

*Observación: En el código que se mostrará a continuación se genera un conjunto de números entre un valor de inicio en este caso 0 y uno final que es 100, especificando el número de puntos entre ellos en el eje x. Para el eje y se grafica una señal seno.*



```python
#LIBRERÍAS
import numpy as np #np será la abreviación para numpy.
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib

#VARIABLES
t = np.linspace(0,10,100) #Lista de valores que puede tomar la variable t. Comienzo, final y número de puntos en el rango.

#FUNCIÓN
def h(t):
  return np.sin(t)

#GRÁFICA
plt.plot(t,h(t)) #Grafica del seno desde 0 hasta 10.
plt.xlabel('t') #Nombre eje x.
plt.ylabel('h(t)') #Nombre eje y.
plt.title('Señal Seno') #Título de la gráfica.
plt.grid(True) #Cuadrícula.
```

**1.2 Señales discretas**

Las señales discretas se definen sobre un conjunto contable del tiempo y pueden modelarse como funciones con dominio en los enteros.

Las señales discretas pueden verse como versiones muestreadas de las señales continuas, como sucede al digitalizar una canción o al escanear una fotografía.

```python
#LIBRERÍAS
import numpy as np #np será la abreviación para numpy.
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib

#VARIABLES
n = np.arange(0, 4*np.pi, 0.4) #Lista de valores que puede tomar la variable n. Comienzo, final y rango entre ellos.
x = np.sin(n) #Seno.

#GRÁFICA
plt.stem(n,x) #Señal discreta.
plt.xlabel('n') #Nombre eje x.
plt.ylabel('x(n)') #Nombre eje y.
plt.title('Señal discreta con un intervalo de muestreo Ts=0.4s') #Título de la gráfica.
plt.grid(True) #Cuadrícula.
```

*Se puede observar en los codigos algo muy interesante y es que la unica diferencia se da en como se dibuja la función X vs Y, ya sea usando stem o plot. Esto es debido a que la computadora tiene memoria finita y es imposible tener señales continuas representadas en ella. Lo que si es importante tener en cuenta es que plot se usa para representar señales que tienen un valor en cada punto infinitesimal de su variable independiente.*

# **2. En función de su variable dependiente / según su rango**

Las señales pueden clasificarse en dos grupos de acuerdo a su rango: en señales de valor continuo y en señales de valor discreto. De forma analoga al caso del dominio, la naturaleza contable o no contable del rango define el tipo de señal.

**2.1 Señal de valor continuo**

Son aquellas que pueden tomar cualquier valor continuo, ya sea en un intervalo finito o infinito. Es decir, su rango es un conjunto continuo, por ejemplo, los reales ℝ o los complejos ℂ.



```python
#LIBRERÍAS
import numpy as np #np será la abreviación para numpy.
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib

#VARIABLES
x = np.arange(0, 4*np.pi, 0.2) #Lista de valores que puede tomar la variable x. Comienzo, final y rango entre ellos.
y = np.sin(x)

#GRÁFICA
fig, axs = plt.subplots(2)
fig.suptitle('Señales de valor continuo')
axs[0].plot(x, y) #Seno de valor continuo en tiempo continuo.
axs[1].stem(x, y) #Seno de valor continuo en tiempo discreto.
```

*Observación: Ambas señales son de valor continuo, pero la primera tiene variable continua y la segunda variable discreta.*

**2.2 Señal de valor discreto**

Las señales de valor discreto son aquellas que sólo pueden tomar valores discretos, es decir valores de un conjunto numerable. También se les puede decir señales cuantizadas o cuantificadas.


```python
#LIBRERÍAS
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib

#VARIABLES
y=[0,0,2,2,4,4,2,2,0,0] #Valores en y.
x=[0,0.5,0.51,1,1.01,2,2.01,2.5,2.51,3] #Valores para señal de tiempo continuo. 
x1 = [0,0.5,0.7,1,1.5,2,2.3,2.5,2.7,3] #Valores para señal en tiempo discreto.

#GRÁFICA
fig, axs = plt.subplots(2)
fig.suptitle('Señales de valor discreto')
axs[0].plot(x, y) #Seno de valor discreto en tiempo continuo.
axs[1].stem(x1, y) #Seno de valor discreto en tiempo discreto.
```

# **3. En función de su máximo**

**3.1 Acotadas o de valor fínito**

Son aquellas que solo toman valores finitos si son reales o de magnitud finita si son complejas. Esto se puede expresar de la siguiente manera: 


acotada.PNG



```python
#LIBRERÍAS
import numpy as np #np será la abreviación para numpy.
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib

#VARIABLES
t = np.arange(0, 4*np.pi, 0.1) #Lista de valores que puede tomar la variable t. Comienzo, final y rango entre ellos.
x = np.cos(t)

#GRÁFICA
plt.plot(t,x) #Grafica del coseno
plt.xlabel('t') #Nombre eje x.
plt.ylabel('x(t)') #Nombre eje y.
plt.title('Señal Coseno') #Título de la gráfica.
plt.grid(True) #Cuadrícula.
```


**3.2 No acotadas o de valor ínfinito**

Son aquellas pueden tomar alguno o varios valores infinitos, tanto en amplitud si son reales o en magnitud si son complejas.

```python
#LIBRERÍAS
import numpy as np #np será la abreviación para numpy.
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib

#VARIABLES
t = np.linspace(0,100,50) #Lista de valores que puede tomar la variable t. Comienzo, final y número de puntos en el rango.
f = np.exp(t) #Función exponencial.

#GRÁFICA
plt.plot(t,f) #Gráfica de la función exponencial.
plt.xlabel('t') #Nombre eje x.
plt.ylabel('f(t)') #Nombre eje y.
plt.title('Función exponencial') #Título de la gráfica.
plt.grid(True) #Cuadrícula.
```

# **4. En función de su duración**

**4.1 Duración finita o soporte compacto**

Son aquellas a las cuales se les puede incluir en un intervalo finito todos sus valores diferentes de cero. 


```python
#LIBRERÍAS
import numpy as np #np será la abreviación para numpy.
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib
from scipy.signal import square, unit_impulse 

#FUNCIÓN
def ventana(t, s, valor):
  x_temp = square(s, 1)
  x = np.array([])
  index = 0
  for value in t:
    value = round(value,1)
    if value in np.around(s, 1):
      x = np.append(x, x_temp[index])
      index += 1
    else:
      x = np.append(x, 0.0)

  x = x * valor
  return x

#VARIABLES
t = np.linspace(-20, 20, 400) #Valores donde se visualizará la gráfica.
s = np.linspace(0, 10, 100, endpoint=False) #Soporte de la ventana.
fx = lambda t, s: ventana(t, s, 4)
x = fx(t,s) #Ventana.

#GRÁFICA
plt.plot(t,x) #Gráfica de la ventana.
plt.xlabel('t') #Nombre eje x.
plt.ylabel('x(t)') #Nombre eje y.
plt.title('Ventana con soporte de 0 a 10') #Título de la gráfica.
plt.grid(True) #Cuadrícula.
```

**4.2 De duración ínfinita o sin soporte compacto**

Son aquellas a las cuales no se les puede incluir en un intervalo
finito todos sus valores diferentes de cero. Como por ejemplo una señal sinc (esta señal se estudiará en detalle en el transcurso del curso), seno o coseno no acotada.

```python
#LIBRERÍAS
import numpy as np #np será la abreviación para numpy.
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib

#VARIABLES
t = np.linspace(-4, 4, 100) #Lista de valores que puede tomar la variable t. Comienzo, final y número de puntos en el rango.
x=np.sinc(t) #Sinc.

#GRÁFICA
plt.plot(t, x) #Gráfica del sinc.
plt.xlabel("t") #Nombre eje x.
plt.ylabel("x(t)") #Nombre eje y.
plt.title("Función Sinc") #Título de la gráfica.
plt.grid(True) #Cuadrícula
```

# Señales especiales


**1.Señal paso, escalón unitario o Heaviside**

paso.PNG

```python
#LIBRERÍAS
import numpy as np #np será la abreviación para numpy.
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib

#VARIABLES
t=np.arange(-2,2,0.01) #Lista de valores que puede tomar la variable t. Comienzo, final y rango entre ellos.
u=(t>=0)*1.0 #Señal paso.

#GRÁFICA
plt.plot(t,u) #Gráfica de la señal paso.
plt.xlabel("t") #Nombre eje x.
plt.ylabel("x(t)") #Nombre eje y.
plt.title("Señal paso") #Título de la gráfica.
plt.grid(True) #Cuadrícula.
```

**Señal impulso, delta de Dirac**

```python
#LIBRERÍAS
import numpy as np #np será la abreviación para numpy.
import matplotlib.pyplot as plt #plt será la abreviación para matplotlib

#VARIABLES
t=np.arange(-200,200,0.5) #Lista de valores que puede tomar la variable t. Comienzo, final y rango entre ellos.
amplitud = 3.0 #Amplitud del impulso.
delta = (t==0)*amplitud #Impulso en t=0.

#GRÁFICA
plt.plot(t,delta) #Gráfica del impulso.
plt.xlabel("t") #Nombre eje x.
plt.ylabel("x(t)") #Nombre eje y.
plt.title("Señal impulso en t=0") #Título de la gráfica.
plt.grid(True) #Cuadrícula.
```
